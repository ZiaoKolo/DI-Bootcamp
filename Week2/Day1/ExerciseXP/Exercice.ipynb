### Catégorisation des colonnes : Quantitative vs Qualitative

#### Définitions:
- **Données Quantitatives** : Numériques, mesurables et permettent des calculs (somme, moyenne, etc.)
  - Exemple: âge, salaire, température, nombre d'enfants
- **Données Qualitatives** : Textuelles, catégorielles, descriptives et non mesurables
  - Exemple: couleur, genre, pays, nom

#### Comment procéder:
1. Examiner le type de données (`dtype`) de chaque colonne
2. Analyser les valeurs uniques et leur contenu
3. Classifier:
   - **Quantitative**: int, float
   - **Qualitative**: object, string, bool, category

```python
# Fonction pour catégoriser les colonnes
def categoriser_columns(dataset, name):
    print(f"ANALYSE: {name}")

    quantitative = []
    qualitative = []
    
    for column in dataset.columns:
        dtype = dataset[column].dtype
        
        # Classifier par type de donnée
        if dtype in ['int64', 'float64', 'int32', 'float32']:
            quantitative.append(column)
        else:
            qualitative.append(column)
    
    print(f"COLONNES QUANTITATIVES ({len(quantitative)}):")
    for col in quantitative:
        print(f"  - {col}: {dataset[col].dtype}")
    
    print(f"COLONNES QUALITATIVES ({len(qualitative)}):")
    for col in qualitative:
        print(f"  - {col}: {dataset[col].dtype}")
    
# Appliquer la fonction à tous les datasets
categoriser_columns(cleaned_data, "clean_dataset.csv")
categoriser_columns(crx_data, "crx.csv")
categoriser_columns(Mental_health_data, "Mental health Depression disorder Data.csv")
categoriser_columns(Time_Americans_Spend_Sleeping, "Time Americans Spend Sleeping.csv")
```

### Exercice 6 - Données structurées ou non structurées

1. **Les rapports financiers d'une entreprise sont stockés dans un fichier Excel** : **données structurées**. Les informations sont organisées en lignes et en colonnes.
2. **Photographies téléchargées sur une plateforme de médias sociaux** : **données non structurées**. Une image ne suit pas un format tabulaire.
3. **Une collection d'articles de presse sur un site web** : **données non structurées**. Le contenu est principalement textuel et n'est pas stocké sous forme de tableau.
4. **Données d'inventaire dans une base de données relationnelle** : **données structurées**. Elles sont organisées selon un schéma précis.
5. **Enregistrements d'entretiens issus d'une étude de marché** : **données non structurées**. Le format audio ou vidéo ne suit pas une structure de table classique.

### Comment raisonner
- Si les données sont organisées dans des tableaux avec des colonnes et des lignes, elles sont généralement **structurées**.
- Si les données sont du texte libre, des images, de l'audio ou de la vidéo, elles sont généralement **non structurées**.

### Exercice 7 - Transformation de données non structurées en données structurées

1. **Une série d'articles de blog sur les expériences de voyage** : on peut les convertir en données structurées en extraisant des champs comme le titre, la date, l'auteur, la destination, les mots-clés et le sentiment du texte. Cela permet de comparer les articles entre eux et de faire une analyse de tendances ou de thèmes récurrents.

2. **Enregistrements audio des appels au service client** : on peut les transformer en texte grâce à la reconnaissance vocale, puis organiser les informations dans un tableau avec l'heure de l'appel, le client, le motif de l'appel, la durée et le niveau de satisfaction. Cette méthode est utile pour analyser les problèmes les plus fréquents et améliorer le service.

3. **Notes manuscrites issues d'une séance de brainstorming** : on peut les numériser avec un outil de reconnaissance de caractères, puis classer les idées dans un tableau selon le thème, la priorité, l'auteur ou le statut. Cela permet de trier les idées plus facilement et de suivre les actions à mettre en place.

4. **Un tutoriel vidéo sur la cuisine** : on peut extraire le texte de la parole, repérer les étapes de la recette, les ingrédients, les temps de cuisson et les actions principales, puis les ranger dans une structure tabulaire. Cela rend le contenu plus facile à rechercher, comparer et réutiliser.

### Méthode générale
- Identifier le contenu principal de la source non structurée.
- Extraire les informations utiles avec des outils adaptés : OCR pour le texte manuscrit, transcription pour l'audio, extraction de texte ou de métadonnées pour les vidéos et les articles.
- Organiser ces informations dans un tableau avec des colonnes claires.
- Vérifier que les données sont cohérentes et comparables avant l'analyse.
