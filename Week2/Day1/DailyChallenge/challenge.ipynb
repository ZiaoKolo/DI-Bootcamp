Étape 1 : Recherche d'un cas pertinent ou d'un article d'actualité
Ce qui a été fait : Nous avons sélectionné et étudié le cas de la multinationale de transport et logistique UPS et de son système d'optimisation d'itinéraires ORION (On-Road Integrated Optimization and Navigation).Pour apporter une perspective sectorielle plus large et actuelle, ce cas principal a été comparé à deux autres leaders du marché :  

Starbucks et son outil d'intelligence géospatiale Atlas utilisé pour l'implantation de ses cafés.  

Rolls-Royce et son système de maintenance prédictive IntelligentEngine pour la gestion de ses moteurs d'avions en temps réel.  

Étape 2 : Analyse du rôle de l'analyse des données (Données, méthodes et décisions)
Ce qui a été fait : L'étude a décortiqué précisément le fonctionnement technique du système ORION d'UPS :

Les données analysées : Le système ingère et traite en continu plus de 250 millions de points de données par jour sont issus de capteurs télématiques IoT installés sur les véhicules (vitesse, freinage, temps de ralentissement), combinés avec les données de trafic en direct, la météo et l'historique des livraisons.  

Les méthodes utilisées : Pour résoudre le problème de itinéraire de véhicules (qui représente un défi mathématique d'une complexité de $6.8\times10^{60}$combinaisons possibles pour un chauffeur effectuant 120 arrêts), ORION combine des heuristiques de réduction d'espace et des métaheuristiques (algorithmes génétiques, recherche Tabu) pour évaluer 200 000 options de trajets par seconde.  

Les résultats et décisions : Le système prescrit aux conducteurs des itinéraires optimisés et dynamiques (par exemple, en entraînant au maximum les virages à gauche coupant le trafic inverse) et intègre des outils comme UPSNav pour identifier l'entrée exacte des zones de livraison.  

Étape 3 : Évaluation de l'impact de l'analyse des données (Scénarios contrefactuels)
Ce qui a été fait : Nous avons mesuré l'impact d'ORION en comparant les résultats actuels à un scénario où l'entreprise n'utilisait pas ces données :

L'impact réel : ORION permet de réduire les trajets de 100 millions de miles par an, économisant 10 millions de gallons de carburant et permettant l'émission de 100 000 tonnes métriques de CO2, tout en générant 300 à 400 millions de dollars d'économies annuelles.  

Sans l'analyse des données (Scénario contrefactuel) : Face à l'explosion de la demande (hausse de près de 20 % du volume de colis entre 2020 et 2023) et à l'augmentation du prix du carburant (qui représente 28 % du budget d'exploitation des flottes de camions en 2024), UPS aurait subi des pertes colossales. Sans ORION, l'entreprise aurait dû étendre physiquement sa flotte pour faire face aux photos de demande (comme la hausse de 15 % constatée lors des fêtes de fin d'année 2024, qui a été absorbée à constante flotte grâce à l'algorithme). Les retards de livraison et le temps d'attente aux intersections auraient fortement dégradé l'efficacité et la satisfaction client.  

Étape 4 : Présentation des conclusions par écrit (Importance de l'analyse)
Ce qui a été fait : Les conclusions du rapport mettent en évidence les principes directeurs pour réussir l'intégration de l'analyse de données dans les décisions modernes d'une entreprise :

La centralisation de la donnée : Connecter et unifier les données de tous les services (opérations, ventes, finance) pour éviter les goulots d'étranglement.  

La conduite du changement et le facteur humain : Le succès d'ORION n'a pas seulement dépendu de l'algorithme, mais de l'accompagnement des chauffeurs. UPS a dû coconcevoir le produit avec eux, adapter ses critères d'évaluation des performances pour récompenser la fidélité à l'itinéraire calculéet partager les bénéfices d'efficacité.  

Le déploiement progressif : L'importance de tester les modèles via des projets pilotes à petite échelle avant de les déployer à l'échelle mondiale.
