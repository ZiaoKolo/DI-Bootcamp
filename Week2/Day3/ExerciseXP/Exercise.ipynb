## Exercice 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

nu = 50
sigma = 10

distribution = norm(nu, sigma)
x = np.linspace(nu - 4 * sigma, nu + 4 * sigma, 200)

plt.plot(x, distribution.pdf(x))
plt.title("Distribution normale (moyenne = 50, écart type = 10)")
plt.xlabel("Valeurs")
plt.ylabel("Densité")
plt.show()
```

## Exercice 4

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)
print(f"Statistique t : {t_stat}, p-value : {p_value}")
```

## Exercice 5

```python
house_sizes =np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

resultat = stats.linregress(house_sizes, house_prices)
print(f"La pente de la régression linéaire est : {resultat.slope}, et l'ordonnée à l'origine est : {resultat.intercept}")

# Estimez le prix d'une maison de 90 mètres carrés.
# Interprétez la signification de la pente dans le contexte des prix de l'immobilier.
prix_estime = resultat.slope * 90 + resultat.intercept
print(f"Le prix estimé d'une maison de 90 mètres carrés est : {prix_estime}")
```

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f"Statistique F : {f_stat}, p-value : {p_value}")
```

## Exercice 8

```python
import pandas as pd
import scipy.stats as stats
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
correlation1, p_value1 = stats.pearsonr(data['age'], data['income'])
correlation2, p_value2 = stats.spearmanr(data["age"], data["income"])

print(f"Corrélation de Pearson : {correlation1}, Valeur p : {p_value1}")
print(f"Corrélation de Spearman : {correlation2}, Valeur p : {p_value2}")
```
